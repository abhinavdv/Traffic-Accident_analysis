# Accident
1. Combined the 3 files we got from API
2. Remove the extra columns and keep only required columns
3. Merge caseyear and st-case (becuase each year has the case number starting for 10001)


```python
df_1 = pd.read_csv("./Accidents Data/Accidents_2010_2013.CSV")
df_2 = pd.read_csv("./Accidents Data/Accidents_2014_2017.CSV")
df_3 = pd.read_csv("./Accidents Data/Accidents_2018_2021.CSV")
```

```python
list(df_1)
```

```python
# List of columns you want to select
columns_to_select = [
    "statename",
    "st_case",
    "caseyear",
    "pernotmvit",
    "ve_total",
    "county",
    "permvit",
    "countyname",
    "city",
    "cityname",
    "month",
    "monthname",
    "day",
    "day_week",
    "day_weekname",
    "year",
    "hour",
    "hourname",
    "minute",
    "minutename",
    "tway_id",
    "routename",
    "func_sysname",
    "latitude",
    "longitud",
    "harm_evname",
    "typ_intname",
    "lgt_cond",
    "lgt_condname",
    "weather",
    "weathername",
    "arr_hour",
    "arr_min",
    "fatals"
]

# Selecting specific columns
selected_columns_df_1 = df_1[columns_to_select]
selected_columns_df_2 = df_2[columns_to_select]
selected_columns_df_3 = df_3[columns_to_select]
```

```python
accidents_data_2010_2021 = pd.concat([selected_columns_df_1, selected_columns_df_2, selected_columns_df_3], ignore_index=True)
```

```python
accidents_data_2010_2021['st_case_per_year'] = accidents_data_2010_2021['caseyear'].astype(str) + '-' + accidents_data_2010_2021['st_case'].astype(str)
```

```python
accidents_data_2010_2021
```

```python
accidents_data_2010_2021.to_csv("AccidentsData_2010-2021.csv")
```

### AccidentsData_2010-2021 is the combination of original files cleaned and formatted

# Vehicles
1. Combined the 4 files we got from API
2. Remove the extra columns and keep only required columns
3. Merge caseyear and st-case (becuase each year has the case number starting for 10001)

```python
df_A = pd.read_csv("./Vehicles Data/Vehicles_2010_2011.CSV")
df_B = pd.read_csv("./Vehicles Data/Vehicles_2012_2014.CSV")
df_C = pd.read_csv("./Vehicles Data/Vehicles_2015_2018.CSV")
df_D = pd.read_csv("./Vehicles Data/Vehicles_2019_2021.CSV")
```

```python
df_A
```

```python
import pandas as pd

# Assuming you already have a DataFrame named df

# List of columns you want to extract
columns_to_extract = [
    "ST_CASE",
    "CASEYEAR",
    "VEH_NO",
    "NUMOCCS",
    "BODY_TYP",
    "BODY_TYPNAME",
    "TRAV_SPNAME",
    "DEFORMED",
    "DEFORMEDNAME",
    "VPICMAKE",
    "VPICMAKENAME"
]

# Convert column names to lowercase
columns_to_extract_lower = [col.lower() for col in columns_to_extract]

# Creating a new DataFrame with only the specified columns
vehicle_df_1 = df_A[columns_to_extract_lower]
vehicle_df_2 = df_B[columns_to_extract_lower]
vehicle_df_3 = df_C[columns_to_extract_lower]
vehicle_df_4 = df_D[columns_to_extract_lower]
```

```python
vehicles_data_2010_2021 = pd.concat([vehicle_df_1, vehicle_df_2, vehicle_df_3, vehicle_df_4], ignore_index=True)
```

```python
vehicles_data_2010_2021['st_case_per_year'] = vehicles_data_2010_2021['caseyear'].astype(str) + '-' + vehicles_data_2010_2021['st_case'].astype(str)
```

```python
vehicles_data_2010_2021
```

```python
vehicles_data_2010_2021.to_csv("VehiclesData_2010-2021.csv")
```
